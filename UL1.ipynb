```python
import math
maxI = 0
maxX = 0
for i in range(1,100):
    x = ((i%6)**2)%7-math.sin(i)
    if maxX<x:
        maxX = x
        maxI = i
        
print maxI
```

```python
import matplotlib.pyplot as plt

values  = []
ivalues  = []
%matplotlib inline
for i in range(-500,1000):
    x = -i**4+1000*(i**3)-20*(i**2)+4*i-6
    values.append(x)
    ivalues.append(i)
plt.plot(ivalues,values)
plt.ylabel('some numbers')
```
